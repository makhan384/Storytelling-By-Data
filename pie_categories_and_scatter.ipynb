```python
import pandas as pd

# Load the dataset to examine its structure
file_path = '/content/drive/MyDrive/New Death Data.csv'
data = pd.read_csv(file_path)

# Display the first few rows and the column names
data.head()
```

```python
data.isnull().values.any()
```

```python
import pandas as pd

# Assuming the first dataset is named data1 and the second dataset is named data2

# Load the datasets

data2 = pd.read_csv("/content/Types of all diseases.csv")

# Merge the datasets on the 'cause_name' column
merged_data = pd.merge(data, data2, on='cause_name', how='inner')

# Display the merged data
print(merged_data)
```

```python
merged_data
```

```python
# Filter the dataset where metric_id is 1
filtered_data = merged_data[merged_data['metric_id'] == 1]

# Group by 'cause_name' and sum the 'val' column to get total death cases
subset_data = filtered_data.groupby('Major types').agg(total=('val', 'sum')).reset_index()

# Display the subset data
print(subset_data.head())
```

```python
import matplotlib.pyplot as plt


# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(subset_data['total'], labels=subset_data['Major types'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of total by cause name')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Aggregate the data by year and cause_name
world_data = data.groupby(['year', 'cause_name'])['val'].sum().reset_index()

# Scatter Plot: Correlation between val and year
scatter_fig = px.scatter(world_data, x='year', y='val', color='cause_name',
                         title='Correlation between Deaths and Year',
                         labels={'Year': 'Year', 'val': 'Number of Deaths', 'cause_name': 'Cause'})
scatter_fig.show()
```
